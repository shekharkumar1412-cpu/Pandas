```python
df=pd.read_csv('New_York_weather_data.csv')
df
```

#**Replacing the values -99999 by NaN value using replace(-99999,value=np.NaN) method**

```python
df1 = df.replace('-99999', np.nan)
df1
```

#**Replacing list with single value if we have more than one more special or wrong value in our file**

```python
new_df = df.replace(['-99999','-88888'], value=0)
new_df
```

#**Replacing per column**

```python
new_df = df.replace({
        'temperature': '-99999',
        'windspeed': '-99999',
        'event': '0'
    }, np.nan)
new_df
```

```python
new_df=df.replace({'event':'0'},value='no event')
new_df
```

#**Replacing by using mapping**

```python
new_df = df.replace({
        -99999: np.nan,     # here we can also use any value like '0 etc
        '0': 'Sunny',
    })
new_df
```

#**Regex(regular expression)**

```python
df=pd.read_csv('New_York_weather_data.csv')
df
```

#In our data file if we encounter any non integer value then we want to remove it then we do as following

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc. then remove non numeric from there
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True)
new_df
```

#**Replacing list with another list**
